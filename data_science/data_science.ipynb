```python
# Import necessary modules
import pandas as pd
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv("/Users/urthor/projects/imc_prosperity/data/bottle-round-1-island-data/combined_prices_round_1.csv", sep=";")

# Inspect the first few rows to verify correct parsing.
df.head(50)
```

```python
# Convert columns to numeric (if not already).
df['day'] = pd.to_numeric(df['day'], errors='coerce')
df['timestamp'] = pd.to_numeric(df['timestamp'], errors='coerce')
df['mid_price'] = pd.to_numeric(df['mid_price'], errors='coerce')
df['profit_and_loss'] = pd.to_numeric(df['profit_and_loss'], errors='coerce')

# Check DataFrame info
df.info()
```

```python

df["day"] = pd.to_numeric(df["day"], errors="coerce")
df["timestamp"] = pd.to_numeric(df["timestamp"], errors="coerce")
df["mid_price"] = pd.to_numeric(df["mid_price"], errors="coerce")

# 2) Create a combined x value, e.g. day * 100000 + timestamp
df["x_axis"] = df["day"] * 100000 + df["timestamp"]

# 3) Separate DataFrames for each product (or you can filter on the fly)
df_resin = df[df["product"] == "RAINFOREST_RESIN"]
df_ink   = df[df["product"] == "SQUID_INK"]
df_kelp  = df[df["product"] == "KELP"]

# 4) Create a figure and the first axes (ax1)
fig, ax1 = plt.subplots(figsize=(10, 5))

# 5) Create two additional “twin” axes that share the same x-axis
ax2 = ax1.twinx()
ax3 = ax1.twinx()

# Shift ax3 to the right so it doesn't overlap ax2’s spine
ax3.spines["right"].set_position(("outward", 60))

# 6) Plot each product on its own axes with thin lines
#    - Ax1 for RAINFOREST_RESIN
lns1 = ax1.plot(
    df_resin["x_axis"], df_resin["mid_price"],
    color="tab:red",
    linewidth=1,
    marker="o",
    markersize=3,
    label="Resin"
)

#    - Ax2 for SQUID_INK
lns2 = ax2.plot(
    df_ink["x_axis"], df_ink["mid_price"],
    color="tab:blue",
    linewidth=1,
    marker="o",
    markersize=3,
    label="Squid Ink"
)

#    - Ax3 for KELP
lns3 = ax3.plot(
    df_kelp["x_axis"], df_kelp["mid_price"],
    color="tab:green",
    linewidth=1,
    marker="o",
    markersize=3,
    label="Kelp"
)

# 7) Label each y-axis
ax1.set_ylabel("Resin Price (red)", color="tab:red")
ax2.set_ylabel("Squid Ink Price (blue)", color="tab:blue")
ax3.set_ylabel("Kelp Price (green)", color="tab:green")

# 8) Label the x-axis
ax1.set_xlabel("Combined Day+Timestamp")

# 9) Optionally combine legends
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines3, labels3 = ax3.get_legend_handles_labels()
ax1.legend(
    lines1 + lines2 + lines3,
    labels1 + labels2 + labels3,
    loc="best"
)

# 10) Show the figure
plt.title("Overlay of Three Products with Multiple Y-Axes")
plt.grid(True)
plt.tight_layout()
plt.show()
```
